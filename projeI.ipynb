```python
import pickle
import os
import re
import pandas as pd
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

custom_stop_words = ['oh', 'a', 'you','let', 'youre', 'get', 'aint', 'say','know', 'yeah', 'lyrics', 'ah', 'ay', 'ooh', 'im', 'dont', 'must', 'la', 'got', 'like', 'need', 'wont', 'ive', 'ohohoh', 'one', 'also', 'may']
stop_words.update(custom_stop_words)

def clean_lyrics(lyrics):
    if not isinstance(lyrics, str):  # Ensure the input is a string
        return ''
    lyrics = re.sub(r'\[.*?\]|\(.*?\)', '', lyrics)
    lyrics = re.sub(r'\d+', '', lyrics)
    lyrics = re.sub(r'[^\w\s]', '', lyrics).replace('\n', ' ')
    lyrics = re.sub(r'\bcontrib\w*\b', '', lyrics, flags=re.IGNORECASE)
    lyrics = lyrics.lower()
    lyrics = ' '.join([word for word in lyrics.split() if word not in stop_words])
    lyrics = ' '.join([word.split('embed')[0] if 'embed' in word else word for word in lyrics.split()])
    return lyrics.strip()

pickle_folder = './pickled_data'
pickle_files = [f for f in os.listdir(pickle_folder) if f.endswith('.pkl')]

for pickle_file in pickle_files:
    pickle_file_path = os.path.join(pickle_folder, pickle_file)

    with open(pickle_file_path, "rb") as file:
        df = pickle.load(file)

    print(f"Veri çerçevesi: {pickle_file}")
    print(df.columns)

    # Drop duplicate columns
    df = df.loc[:, ~df.columns.duplicated()]

    if 'lyrics' in df.columns:
        df['clean_lyrics'] = df['lyrics'].apply(clean_lyrics)

    with open(pickle_file_path, "wb") as file:
        pickle.dump(df, file)

    print(f"Temizlenmiş {pickle_file} başarıyla kaydedildi!")
```

```python
import os
import pickle
import torch
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("cardiffnlp/twitter-roberta-base-sentiment-latest")
model = AutoModelForSequenceClassification.from_pretrained("cardiffnlp/twitter-roberta-base-sentiment-latest")

pickle_folder = './pickled_data'

pickle_files = [f for f in os.listdir(pickle_folder) if f.endswith('.pkl')]

def get_sentiment(lyrics):
    inputs = tokenizer(lyrics, return_tensors="pt", truncation=True, padding=True, max_length=512)
    
    with torch.no_grad():
        outputs = model(**inputs)
        logits = outputs.logits

    probs = torch.nn.functional.softmax(logits, dim=-1)
    
    labels = ['negative', 'neutral', 'positive']

    sentiment_score, sentiment_label_idx = torch.max(probs, dim=-1)
    sentiment_label = labels[sentiment_label_idx.item()]
    
    return sentiment_label, sentiment_score.item()

for pickle_file in pickle_files:
    pickle_file_path = os.path.join(pickle_folder, pickle_file)
    
    with open(pickle_file_path, "rb") as file:
        df = pickle.load(file)
    
    if 'clean_lyrics' in df.columns:
        lyrics_analysis = []
        
        for lyrics in df.clean_lyrics:
            sentiment_label, sentiment_score = get_sentiment(lyrics)
            lyrics_analysis.append({'lyrics': lyrics, 'sentiment': sentiment_label, 'score': sentiment_score})
        
        lyrics_analysis_df = pd.DataFrame(lyrics_analysis)
        
        df = pd.concat([df, lyrics_analysis_df], axis=1)
        
        with open(pickle_file_path, "wb") as file:
            pickle.dump(df, file)
        
        print(f"{pickle_file} başarıyla işlendi ve kaydedildi!")
```

```python
import os
import pickle
import pandas as pd
import torch

def analyze_sentiment(text, tokenizer, model):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=512)
    with torch.no_grad():
        outputs = model(**inputs)
        logits = outputs.logits
        sentiment = torch.argmax(logits, dim=1).item()
        return sentiment

def get_sentiment_label(sentiment):
    if sentiment == 0:
        return 'negative'
    elif sentiment == 1:
        return 'neutral'
    else:
        return 'positive'

pickle_folder = './pickled_data/'  

pickle_files = [
    '1960lar.pkl', '1980ler.pkl', '1990lar.pkl', '2000_2010lar.pkl', '2020ler.pkl'
]

sentiment_by_decade = {}

for pickle_file in pickle_files:
    pickle_file_path = os.path.join(pickle_folder, pickle_file)
    
    with open(pickle_file_path, "rb") as file:
        df = pickle.load(file)

    if 'clean_lyrics' in df.columns:
        sentiments = []

        for lyrics in df['clean_lyrics']:
            lyrics_parts = [lyrics[i:i+512] for i in range(0, len(lyrics), 512)] 

            sentiment_labels = []
            for part in lyrics_parts:
                sentiment = analyze_sentiment(part, tokenizer, model)
                sentiment_label = get_sentiment_label(sentiment)
                sentiment_labels.append(sentiment_label)

            majority_sentiment = max(set(sentiment_labels), key=sentiment_labels.count)
            sentiments.append(majority_sentiment)

        sentiment_by_decade[pickle_file] = pd.Series(sentiments).value_counts()

for decade, sentiment_counts in sentiment_by_decade.items():
    print(f"\nSentiment analysis for {decade}:")
    print(sentiment_counts)
```

```python
import os
import pickle
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import re

custom_stop_words = ['oh', 'a', 'you', 'know', 'cause', 'let','get','say', 'might', 'yeah', 'lyrics', 'ah', 'ay', 'ooh', 'im', 'dont', 'must', 
                     'la', 'got', 'like', 'need', 'wont', 'ive', 'ohohoh', 'one', 'also', 'may']

pickle_folder = './pickled_data/' 

pickle_files = [
    '1960lar.pkl', '1980ler.pkl', '1990lar.pkl', '2000_2010lar.pkl', '2020ler.pkl'
]

word_frequencies_by_decade = {}

def clean_text(text, stop_words):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    words = text.split()
    filtered_words = [word for word in words if word not in stop_words]
    return filtered_words

for pickle_file in pickle_files:
    pickle_file_path = os.path.join(pickle_folder, pickle_file)
    
    with open(pickle_file_path, "rb") as file:
        df = pickle.load(file)

    if 'clean_lyrics' in df.columns:
        lyrics = df['clean_lyrics'].tolist()

        all_lyrics = " ".join(lyrics)

        cleaned_words = clean_text(all_lyrics, custom_stop_words)

        word_frequencies = Counter(cleaned_words)

        word_frequencies_by_decade[pickle_file] = word_frequencies

for decade, word_frequencies in word_frequencies_by_decade.items():
    top_words = word_frequencies.most_common(10)
    words, frequencies = zip(*top_words)
    
    plt.figure(figsize=(10, 6))
    
    plt.barh(words, frequencies, color='skyblue')

    plt.title(f'Top 10 Words in Lyrics for {decade}', fontsize=16)
    plt.xlabel('Frequency', fontsize=14)
    plt.ylabel('Words', fontsize=14)
    
    plt.tight_layout()
    plt.show()
```

```python
import os
import pickle
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import re

keywords = {
    'love': ['love', 'girl', 'baby'],
    'curse': ['fuck', 'shit', 'ass', 'bitch'],
    'war': ['fight', 'war', 'sword', 'scream']
}

pickle_folder = './pickled_data/'  

pickle_files = [
    '1960lar.pkl', '1980ler.pkl', '1990lar.pkl', '2000_2010lar.pkl', '2020ler.pkl'
]

keywords_frequency_by_decade = {key: {pickle_file: 0 for pickle_file in pickle_files} for key in keywords}

def clean_text(text, stop_words):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    words = text.split()
    filtered_words = [word for word in words if word not in stop_words]
    return filtered_words

for pickle_file in pickle_files:
    pickle_file_path = os.path.join(pickle_folder, pickle_file)
    
    with open(pickle_file_path, "rb") as file:
        df = pickle.load(file)

    if 'clean_lyrics' in df.columns:
        lyrics = df['clean_lyrics'].tolist()

        all_lyrics = " ".join(lyrics)

        cleaned_words = clean_text(all_lyrics, custom_stop_words)

        for label, keyword_list in keywords.items():
            count = 0
            for keyword in keyword_list:
                count += cleaned_words.count(keyword)
            keywords_frequency_by_decade[label][pickle_file] = count

for label, decade_data in keywords_frequency_by_decade.items():
    decades = list(decade_data.keys())
    counts = list(decade_data.values())
    
    plt.figure(figsize=(10, 6))
    
    plt.barh(decades, counts, color='orange')

    plt.title(f'{label.capitalize()} Word Group Frequency Over Decades', fontsize=16)
    plt.xlabel('Frequency', fontsize=14)
    plt.ylabel('Decades', fontsize=14)
    
    plt.tight_layout()
    plt.show()
```
