```python
import json
import pandas as pd
import os

# JSON dosyalarının bulunduğu klasör
json_folder = "./"  # Dosyalar mevcut çalışma dizinindeyse

# JSON dosyalarının listesini alın
json_files = [file for file in os.listdir(json_folder) if file.endswith(".json")]

# Her dosya için DataFrame'leri saklamak için bir sözlük
dataframes = {}

# Dosyaları işleme
for json_file in json_files:
    with open(json_file, "r", encoding="utf-8") as file:
        data = json.load(file)
    
    master_list = []
    
    # Şarkı verilerini işleme
    for row in data.get('songs', []):  # "songs" anahtarını kontrol et
        temp_dict = {
            "title": row.get("title"),
            "release_date": row.get("release_date"),
            "album": row.get("album", {}).get("name") if row.get("album") else None,
            "lyrics": row.get("lyrics"),
        }
        master_list.append(temp_dict)
    
    # DataFrame oluştur ve sözlüğe ekle
    df = pd.DataFrame(master_list)
    dataframes[json_file.replace(".json", "")] = df  # Dosya adını anahtar olarak kullan
    

# Örnek: Tüm DataFrame'leri ayrı CSV'lere kaydetmek isterseniz
for name, df in dataframes.items():
    df.to_csv(f"{name}.csv", index=False, encoding="utf-8")
```

```python
import pickle
import os
import re
import nltk
from nltk.corpus import stopwords

# NLTK stopwords'lerini indir
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Özelleştirilmiş stopwords listesi
custom_stop_words = ['oh', 'a', 'you', 'yeah', 'lyrics', 'ah', 'ay', 'ooh', 'im', 'dont', 'must', 'la', 'got', 'like', 'need', 'wont', 'ive', 'ohohoh', 'one', 'also', 'may']
stop_words.update(custom_stop_words)

# Temizlik fonksiyonu
def clean_lyrics(lyrics):
    # Parantez içi metinleri kaldır
    lyrics = re.sub(r'\[.*?\]|\(.*?\)', '', lyrics)
    # Sayıları kaldır
    lyrics = re.sub(r'\d+', '', lyrics)
    # Noktalama işaretlerini ve satır sonlarını kaldır
    lyrics = re.sub(r'[^\w\s]', '', lyrics).replace('\n', ' ')
    
    # 'contributorstranslations' ve onu takip eden dillerle ilgili metinleri temizle
    lyrics = re.sub(r'\bcontrib\w*\b', '', lyrics, flags=re.IGNORECASE)
    lyrics = lyrics.lower()
    
    # Stop words'leri kaldır
    lyrics = ' '.join([word for word in lyrics.split() if word not in stop_words])
    
    # 'embed' içeren kelimeleri ayır ve sadece 'embed' öncesini al
    lyrics = ' '.join([word.split('embed')[0] if 'embed' in word else word for word in lyrics.split()])
    
    return lyrics.strip()

# Pickle dosyasının bulunduğu klasör
pickle_folder = './pickled_data'

# Pickle dosyalarını al
pickle_files = [f for f in os.listdir(pickle_folder) if f.endswith('.pkl')]

# Her pickle dosyasını işleme
for pickle_file in pickle_files:
    pickle_file_path = os.path.join(pickle_folder, pickle_file)
    
    # Pickle dosyasını yükle
    with open(pickle_file_path, "rb") as file:
        df = pickle.load(file)
    
    # Sütunları kontrol et
    print(f"Veri çerçevesi: {pickle_file}")
    print(df.columns)  # Sütun adlarını kontrol et
    
    # Temizlik fonksiyonunu lyrics sütununa uygula
    if 'lyrics' in df.columns:
        df['clean_lyrics'] = df['lyrics'].apply(clean_lyrics)
    
    # İşlenmiş veriyi tekrar pickle olarak kaydet
    with open(pickle_file_path, "wb") as file:
        pickle.dump(df, file)

    print(f"Temizlenmiş {pickle_file} başarıyla kaydedildi!") 
```

```python
import os
import pickle
import requests
import pandas as pd

# Hugging Face model ve API URL
model = "cardiffnlp/twitter-roberta-base-sentiment-latest"
hf_token = "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"
API_URL = "https://api-inference.huggingface.co/models/" + model
headers = {"Authorization": "Bearer %s" % (hf_token)}

# Duygu analiz fonksiyonu
def analysis(data):
    payload = dict(inputs=data, options=dict(wait_for_model=True))
    response = requests.post(API_URL, headers=headers, json=payload)
    return response.json()

# Pickle dosyalarının bulunduğu klasör
pickle_folder = './pickled_data'

# Pickle dosyalarını al
pickle_files = [f for f in os.listdir(pickle_folder) if f.endswith('.pkl')]

# Her pickle dosyasını işleme
for pickle_file in pickle_files:
    pickle_file_path = os.path.join(pickle_folder, pickle_file)
    
    # Pickle dosyasını yükle
    with open(pickle_file_path, "rb") as file:
        df = pickle.load(file)
    
    # Eğer 'clean_lyrics' sütunu varsa
    if 'clean_lyrics' in df.columns:
        lyrics_analysis = []
        
        # Şarkı sözlerini analiz et
        for lyrics in df['clean_lyrics']:
            try:
                sentiment_result = analysis(lyrics)[0]  # Duygu analizinin ilk sonucu
                sentiment_probabilities = {label['label']: label['score'] for label in sentiment_result}
                lyrics_analysis.append({'lyrics': lyrics, **sentiment_probabilities})
            except Exception as e:
        
        # Analiz sonuçlarını DataFrame'e ekle
        lyrics_analysis_df = pd.DataFrame(lyrics_analysis)
        
        # DataFrame'i orijinal DataFrame'e ekle
        df = pd.concat([df, lyrics_analysis_df], axis=1)
        
        # İşlenmiş veriyi tekrar pickle olarak kaydet
        with open(pickle_file_path, "wb") as file:
            pickle.dump(df, file)
        
```
