```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

f=[15,60,40,25]
s=[10,15,20,60]
```

```python
plt.plot(f,label="first student books")
plt.plot(s,label="second student books")
plt.title("books students read")
plt.xlabel("months")
plt.ylabel("books")
plt.legend(bbox_to_anchor=(1.02,1), borderaxespad=2)
a= plt.figure()
```
