```python
import json
import pandas as pd
import os

# JSON dosyalarının bulunduğu klasör
json_folder = "./json_files"  # JSON dosyalarının bulunduğu klasör
csv_folder = "./csv_filess"  # CSV dosyalarının kaydedileceği klasör

# CSV dosyalarının klasörünü oluştur (eğer yoksa)
os.makedirs(csv_folder, exist_ok=True)

# JSON dosyalarının listesini alın
json_files = [file for file in os.listdir(json_folder) if file.endswith(".json")]

# Her bir dosya için DataFrame'leri saklayacağımız bir sözlük
dataframes = {}

# Dosyaları işleme
for json_file in json_files:
    json_file_path = os.path.join(json_folder, json_file)  # Tam dosya yolu oluştur
    with open(json_file_path, "r", encoding="utf-8") as file:
        data = json.load(file)
    
    master_list = []
    
    # Eğer data bir liste ise, doğrudan liste elemanlarını işle
    if isinstance(data, list):
        songs = data
    else:
        songs = data.get('songs', [])
    
    # Şarkı verilerini işleme
    for row in songs:
        temp_dict = {
            "title": row.get("title"),
            "release_date": row.get("release_date"),
            "album": row.get("album", {}).get("name") if row.get("album") else None,
            "lyrics": row.get("lyrics"),
        }
        master_list.append(temp_dict)
    
    # DataFrame oluştur ve sözlüğe kaydet
    df = pd.DataFrame(master_list)
    dataframes[json_file] = df

    # DataFrame'i CSV olarak kaydet
    csv_file = os.path.join(csv_folder, json_file.replace(".json", ".csv"))  # CSV dosyasını csv_folder içine kaydet
    df.to_csv(csv_file, index=False, encoding="utf-8")
    print(f"{csv_file} kaydedildi.")

print("Tüm JSON dosyaları işlendi.")
```

```python
import json
import pandas as pd
import os

# Dönem JSON dosyalarının listesi
donem_json_dosyalar = [
    "1960lar.json",
    "1980ler.json",
    "1990lar.json",
    "2000_2010lar.json",
    "2020ler.json",
]

# JSON'dan CSV'ye dönüştürme işlemi
for json_file in donem_json_dosyalar:
    try:
        # JSON dosyasının tam yolunu oluştur
        json_file_path = os.path.join(json_folder, json_file)
        
        # JSON dosyasını yükle
        with open(json_file_path, "r", encoding="utf-8") as file:
            data = json.load(file)
        
        # JSON'daki sanatçı ve şarkı bilgilerini ayır
        master_list = []
        for artist in data.get("artists", []):
            artist_name = artist.get("name", "Bilinmeyen Sanatçı")
            for song in artist.get("songs", []):
                temp_dict = {
                    "artist_name": artist_name,
                    "song_title": song.get("title"),
                    "release_date": song.get("release_date"),
                    "album": song.get("album", {}).get("name") if song.get("album") else None,
                    "lyrics": song.get("lyrics"),
                }
                master_list.append(temp_dict)
        
        # DataFrame oluştur ve CSV'ye kaydet
        df = pd.DataFrame(master_list)
        csv_file_name = os.path.join(csv_folder, json_file.replace(".json", ".csv"))  # CSV dosyasının tam yolu
        df.to_csv(csv_file_name, index=False, encoding="utf-8-sig")  # UTF-8 Sig ile Türkçe desteği
        print(f"{csv_file_name} başarıyla kaydedildi.")
    
    except Exception as e:
        print(f"{json_file} işlenirken bir hata oluştu: {type(e).__name__} - {e}")
```

```python
import pickle
import os
import re
import nltk
from nltk.corpus import stopwords

# NLTK stopwords'lerini indir
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Özelleştirilmiş stopwords listesi
custom_stop_words = ['oh', 'a', 'you', 'yeah', 'lyrics', 'ah', 'ay', 'ooh', 'im', 'dont', 'must', 'la', 'got', 'like', 'need', 'wont', 'ive', 'ohohoh', 'one', 'also', 'may']
stop_words.update(custom_stop_words)

# Temizlik fonksiyonu
def clean_lyrics(lyrics):
    # Parantez içi metinleri kaldır
    lyrics = re.sub(r'\[.*?\]|\(.*?\)', '', lyrics)
    # Sayıları kaldır
    lyrics = re.sub(r'\d+', '', lyrics)
    # Noktalama işaretlerini ve satır sonlarını kaldır
    lyrics = re.sub(r'[^\w\s]', '', lyrics).replace('\n', ' ')
    
    # 'contributorstranslations' ve onu takip eden dillerle ilgili metinleri temizle
    lyrics = re.sub(r'\bcontrib\w*\b', '', lyrics, flags=re.IGNORECASE)
    lyrics = lyrics.lower()
    
    # Stop words'leri kaldır
    lyrics = ' '.join([word for word in lyrics.split() if word not in stop_words])
    
    # 'embed' içeren kelimeleri ayır ve sadece 'embed' öncesini al
    lyrics = ' '.join([word.split('embed')[0] if 'embed' in word else word for word in lyrics.split()])
    
    return lyrics.strip()

# Pickle dosyasının bulunduğu klasör
pickle_folder = './pickled_data'

# Pickle dosyalarını al
pickle_files = [f for f in os.listdir(pickle_folder) if f.endswith('.pkl')]

# Her pickle dosyasını işleme
for pickle_file in pickle_files:
    pickle_file_path = os.path.join(pickle_folder, pickle_file)
    
    # Pickle dosyasını yükle
    with open(pickle_file_path, "rb") as file:
        df = pickle.load(file)
    
    # Sütunları kontrol et
    print(f"Veri çerçevesi: {pickle_file}")
    print(df.columns)  # Sütun adlarını kontrol et
    
    # Temizlik fonksiyonunu lyrics sütununa uygula
    if 'lyrics' in df.columns:
        df['clean_lyrics'] = df['lyrics'].apply(clean_lyrics)
    
    # İşlenmiş veriyi tekrar pickle olarak kaydet
    with open(pickle_file_path, "wb") as file:
        pickle.dump(df, file)

    print(f"Temizlenmiş {pickle_file} başarıyla kaydedildi!") 
```

```python
import os
import pickle
import torch
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# Model ve tokenizer'ı yükleyin
tokenizer = AutoTokenizer.from_pretrained("cardiffnlp/twitter-roberta-base-sentiment-latest")
model = AutoModelForSequenceClassification.from_pretrained("cardiffnlp/twitter-roberta-base-sentiment-latest")

# Pickle dosyalarının bulunduğu klasör
pickle_folder = './pickled_data'

# Pickle dosyalarını al
pickle_files = [f for f in os.listdir(pickle_folder) if f.endswith('.pkl')]

# Duygu analiz fonksiyonu
def get_sentiment(lyrics):
    # Şarkı sözünü token'lara ayırın
    inputs = tokenizer(lyrics, return_tensors="pt", truncation=True, padding=True, max_length=512)
    
    # Modelin tahminini al
    with torch.no_grad():
        outputs = model(**inputs)
        logits = outputs.logits

    # Sonuçları yumuşatarak (softmax) sınıf olasılıklarını hesapla
    probs = torch.nn.functional.softmax(logits, dim=-1)
    
    # Etiketleri al
    labels = ['negative', 'neutral', 'positive']

    # En yüksek olasılığı ve etiketini döndür
    sentiment_score, sentiment_label_idx = torch.max(probs, dim=-1)
    sentiment_label = labels[sentiment_label_idx.item()]
    
    return sentiment_label, sentiment_score.item()

# Her pickle dosyasını işleme
for pickle_file in pickle_files:
    pickle_file_path = os.path.join(pickle_folder, pickle_file)
    
    # Pickle dosyasını yükle
    with open(pickle_file_path, "rb") as file:
        df = pickle.load(file)
    
    # Eğer 'clean_lyrics' sütunu varsa
    if 'clean_lyrics' in df.columns:
        lyrics_analysis = []
        
        # Şarkı sözlerini analiz et
        for lyrics in df.clean_lyrics:
            sentiment_label, sentiment_score = get_sentiment(lyrics)
            lyrics_analysis.append({'lyrics': lyrics, 'sentiment': sentiment_label, 'score': sentiment_score})
        
        # Analiz sonuçlarını DataFrame'e ekle
        lyrics_analysis_df = pd.DataFrame(lyrics_analysis)
        
        # DataFrame'i orijinal DataFrame'e ekle
        df = pd.concat([df, lyrics_analysis_df], axis=1)
        
        # İşlenmiş veriyi tekrar pickle olarak kaydet
        with open(pickle_file_path, "wb") as file:
            pickle.dump(df, file)
        
        print(f"{pickle_file} başarıyla işlendi ve kaydedildi!")
```

```python
import os
import pickle
import pandas as pd
import torch

# Duygu analizi fonksiyonu (model ve tokenizer zaten yüklü)
def analyze_sentiment(text, tokenizer, model):
    # Veriyi tokenize et
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=512)
    with torch.no_grad():
        # Modeli çalıştır
        outputs = model(**inputs)
        logits = outputs.logits
        # Duygu etiketini al (0=negatif, 1=nötr, 2=pozitif)
        sentiment = torch.argmax(logits, dim=1).item()
        return sentiment

# Duygu etiketlerine göre sonucu döndüren fonksiyon
def get_sentiment_label(sentiment):
    if sentiment == 0:
        return 'negative'
    elif sentiment == 1:
        return 'neutral'
    else:
        return 'positive'

# Dosyaların bulunduğu klasör
pickle_folder = './pickled_data/'  # Kendi dosya yolunuzu buraya ekleyin

# Dönemlere göre analiz yapacağımız dosyaların isimleri
pickle_files = [
    '1960lar.pkl', '1980ler.pkl', '1990lar.pkl', '2000_2010lar.pkl', '2020ler.pkl'
]

# Duygu dağılımlarını saklayacağımız bir dictionary
sentiment_by_decade = {}

# Her dosyayı işle
for pickle_file in pickle_files:
    pickle_file_path = os.path.join(pickle_folder, pickle_file)
    
    # Dosyayı yükle
    with open(pickle_file_path, "rb") as file:
        df = pickle.load(file)

    # Eğer 'clean_lyrics' sütunu varsa, şarkı sözlerini analiz et
    if 'clean_lyrics' in df.columns:
        sentiments = []

        # Her şarkı sözünü analiz et
        for lyrics in df['clean_lyrics']:
            # Şarkı sözlerini parçalara böl
            lyrics_parts = [lyrics[i:i+512] for i in range(0, len(lyrics), 512)]  # 512 token'lık parçalara böldük

            sentiment_labels = []
            for part in lyrics_parts:
                sentiment = analyze_sentiment(part, tokenizer, model)
                sentiment_label = get_sentiment_label(sentiment)
                sentiment_labels.append(sentiment_label)

            # Parçaların çoğunluğuna göre sonucu al
            majority_sentiment = max(set(sentiment_labels), key=sentiment_labels.count)
            sentiments.append(majority_sentiment)

        # Duyguları dosya adına göre sakla
        sentiment_by_decade[pickle_file] = pd.Series(sentiments).value_counts()

# Sonuçları yazdır
for decade, sentiment_counts in sentiment_by_decade.items():
    print(f"\nSentiment analysis for {decade}:")
    print(sentiment_counts)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Önceden elde edilen duygu analizi sonuçları
sentiment_data = {
    "1960lar": {"positive": 15, "neutral": 33, "negative": 25},
    "1980ler": {"positive": 30, "neutral": 27, "negative": 23},
    "1990lar": {"positive": 12, "neutral": 22, "negative": 26},
    "2000_2010lar": {"positive": 18, "neutral": 19, "negative": 39},
    "2020ler": {"positive": 4, "neutral": 21, "negative": 34},
}

# Veriyi pandas DataFrame'e dönüştür
sentiment_df = pd.DataFrame(sentiment_data).T

# Grafik oluşturma
plt.figure(figsize=(10, 6))

# Her bir duygu türünü çiz
plt.plot(sentiment_df.index, sentiment_df['positive'], label='Positive', marker='o', color="purple")
plt.plot(sentiment_df.index, sentiment_df['neutral'], label='Neutral', marker='o', color="orange")
plt.plot(sentiment_df.index, sentiment_df['negative'], label='Negative', marker='o', color="blue")

# Başlık ve etiketler
plt.title('Sentiment Analysis for Different Decades', fontsize=16)
plt.xlabel('Decades', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.legend()

# Grafik göster
plt.grid(True)
plt.show()
```

```python
import os
import pickle
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import re

# Dosyaların bulunduğu klasör
pickle_folder = './pickled_data/'  # Kendi dosya yolunuzu buraya ekleyin

# Dönemlere göre analiz yapacağımız dosyaların isimleri
pickle_files = [
    '1960lar.pkl', '1980ler.pkl', '1990lar.pkl', '2000_2010lar.pkl', '2020ler.pkl'
]

# Her dosya için kelime sıklıklarını tutacağımız bir dictionary
word_frequencies_by_decade = {}

# Kelime temizleme fonksiyonu
def clean_text(text):
    # Küçük harfe dönüştürme, noktalama işaretlerinden arındırma
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    return text

# Her dosyayı işle
for pickle_file in pickle_files:
    pickle_file_path = os.path.join(pickle_folder, pickle_file)
    
    # Dosyayı yükle
    with open(pickle_file_path, "rb") as file:
        df = pickle.load(file)

    # Eğer 'clean_lyrics' sütunu varsa, şarkı sözlerini analiz et
    if 'clean_lyrics' in df.columns:
        lyrics = df['clean_lyrics'].tolist()

        # Şarkı sözlerini birleştir
        all_lyrics = " ".join(lyrics)

        # Temizle ve kelimeleri ayır
        cleaned_lyrics = clean_text(all_lyrics)
        words = cleaned_lyrics.split()

        # Kelime sıklığını hesapla
        word_frequencies = Counter(words)

        # Dönem ismi ile kaydet
        word_frequencies_by_decade[pickle_file] = word_frequencies

# Her dönemi için ayrı grafik çizme
for decade, word_frequencies in word_frequencies_by_decade.items():
    # En sık 10 kelimeyi al
    top_words = word_frequencies.most_common(10)
    words, frequencies = zip(*top_words)
    
    # Yeni bir figure oluştur
    plt.figure(figsize=(10, 6))
    
    # Bar chart çiz
    plt.barh(words, frequencies, color='skyblue')

    # Grafik başlıkları ve etiketler
    plt.title(f'Top 10 Words in Lyrics for {decade}', fontsize=16)
    plt.xlabel('Frequency', fontsize=14)
    plt.ylabel('Words', fontsize=14)
    
    # Grafik göster
    plt.tight_layout()
    plt.show()
```
